# Introduction

This notebook provides an in-depth analysis of our multi-modal garbage classification model.
I combine a pretrained MobileNetV2 for image features and DistilBERT for text features.
My goal is to classify garbage items into four categories (Black, Blue, Green, TTR) by leveraging both modalities.

We'll explore:
- Data distribution and potential class imbalance
- Model performance metrics (accuracy, precision, recall, F1-score)
- Confusion matrix 
- Visualization of misclassified samples


# 1. Notebook Setup

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Import dataset loader and model architecture
from dataset_loader import test_loader
from model_architecture import MultiModalGarbageClassifier

# Set device to GPU (if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

# 2. Data Exploration

```python
import numpy as np
import matplotlib.pyplot as plt
from dataset_loader import train_loader, val_loader, test_loader

all_labels = [s[2] for s in train_loader.dataset.samples]
unique, counts = np.unique(all_labels, return_counts=True)

plt.bar(unique, counts)
plt.xticks(unique, ["Black","Blue","Green","TTR"])
plt.title("Class Distribution in Training Set")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

print("Class counts:", dict(zip(unique, counts)))
```

```python
# Compute the proportion of each class
total_samples = sum(counts)
class_proportions = counts / total_samples

print("Class Distribution Percentages:")
for label, count, proportion in zip(["Black", "Blue", "Green", "TTR"], counts, class_proportions):
    print(f"{label}: {count} samples ({proportion*100:.2f}%)")

# Compute imbalance ratio (max class / min class)
imbalance_ratio = max(counts) / min(counts)
print(f"\nImbalance Ratio (Max Class / Min Class): {imbalance_ratio:.2f}")

# Compute Shannon entropy for dataset balance
entropy = -np.sum(class_proportions * np.log2(class_proportions))
print(f"Class Distribution Entropy: {entropy:.2f} (Closer to 2 means balanced, closer to 0 means imbalanced)")
```

The bar chart shows the number of samples for each garbage class in our **training dataset**:

- **Black**: 2368 samples (~20.36%)
- **Blue**: 4842 samples (~41.64%)
- **Green**: 2317 samples (~19.92%)
- **TTR**: 2102 samples (~18.08%)

**Imbalance Ratio (Max Class / Min Class):** 2.30  
This means the largest class (Blue) has **2.3×** as many samples as the smallest class (TTR).

**Class Distribution Entropy:** 1.90  
(Closer to 2 indicates better balance, closer to 0 indicates heavy imbalance.)

Overall, the dataset is moderately imbalanced, which can affect model training and performance.

# 3. Model Architecture

I use two pretrained networks:
- **MobileNetV2** for image feature extraction, outputting a 256-dimensional feature vector.
- **DistilBERT** for text feature extraction, reduced from 768 → 256 dimensions.

I then:
1. **Normalize** each 256-d feature vector with LayerNorm to align scales.
2. **Concatenate** them for a combined 512-d feature.
3. Pass through a fusion layer (512 → 512) and then a classification layer (512 → 4).

# 4. Model Evaluation & Metrics

## 4.1. Load the Best Model

```python
num_classes = 4 
model = MultiModalGarbageClassifier(num_classes=num_classes)
model.load_state_dict(torch.load("best_model4.pth", map_location=device, weights_only=True))
model.to(device)
model.eval()
```

# 4.2. Run Inference on the Test Set

```python
all_preds = []
all_labels = []
all_images = []
all_texts = []

with torch.no_grad():
    for images, texts, labels in test_loader:
        images = images.to(device)
        # Tokenize texts on the fly using DistilBERT tokenizer
        from transformers import DistilBertTokenizer
        tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
        encoding = tokenizer(list(texts), padding="longest", truncation=True, return_tensors="pt")
        encoding = {k: v.to(device) for k, v in encoding.items()}

        outputs = model(images, encoding)
        _, preds = torch.max(outputs, 1)
        
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_images.extend(images.cpu()) 
        all_texts.extend(texts)
```

## 4.3. Compute and Display Metrics

```python
print(classification_report(all_labels, all_preds, target_names=["Black", "Blue", "Green", "TTR"]))
```

**Overall Accuracy:** 86%  
**Macro Avg F1:** ~0.85  
**Weighted Avg F1:** ~0.86  

- **Green** has the **highest** recall (94%), meaning the model captures most Green items correctly.
- **Black** has the **lowest** recall (74%), showing it’s more frequently misclassified (often as TTR or Blue).
- The dataset imbalance (especially for TTR and Black) likely impacts these metrics.

**Key Takeaway**: The model performs well overall (86% accuracy), but **Black** and **TTR** classes need further improvement. 

# 4.4. Plot a Confusion Matrix

```python
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=["Black", "Blue", "Green", "TTR"],
            yticklabels=["Black", "Blue", "Green", "TTR"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

Below is the confusion matrix for our **test set** predictions:

- **Black (row 1)**: 104 samples misclassified as Blue, 60 as TTR.
- **Blue (row 2)**: Highest correct classification (991). 
- **Green (row 3)**: Very few misclassifications (mostly correct).
- **TTR (row 4)**: 69 samples confused as Black, 71 as Blue.

We can see **notable confusion** between **Black ↔ TTR** and **Blue ↔ TTR**, likely influenced by the dataset's class imbalance. Improving the representation of underrepresented classes could help reduce these confusions.


# 5. Misclassifications & Visualizations

```python
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import torch

def unnormalize(img, mean, std):
    """
    Unnormalizes an image tensor using the provided mean and std.
    img: Tensor of shape [C, H, W]
    mean, std: Iterable of length 3.
    Returns the unnormalized image tensor.
    """
    # Clone the tensor to avoid modifying the original
    img = img.clone()
    for t, m, s in zip(img, mean, std):
        t.mul_(s).add_(m)
    return img

class_names = {0: "Black", 1: "Blue", 2: "Green", 3: "TTR"}

misclassified_indices = [i for i, (pred, true) in enumerate(zip(all_preds, all_labels)) if pred != true]

print(f"Total misclassifications: {len(misclassified_indices)}")

# Define the ImageNet normalization parameters
mean = torch.tensor([0.485, 0.456, 0.406])
std = torch.tensor([0.229, 0.224, 0.225])

# Visualize a few misclassified samples:
num_to_display = 5
for i in misclassified_indices[:num_to_display]:
    img = all_images[i]
    # Unnormalize the image
    unnorm_img = unnormalize(img, mean, std)
    # Clamp the values to [0, 1] and convert to NumPy
    np_img = unnorm_img.clamp(0, 1).permute(1, 2, 0).numpy()
    
    plt.figure()
    plt.imshow(np_img)
    plt.title(f"True: {class_names[all_labels[i]]} | Predicted: {class_names[all_preds[i]]}\nDescription: {all_texts[i]}")
    plt.axis("off")
    plt.show()
```

```python
# Compute misclassification rates per class
misclass_counts = np.zeros(4)
for true, pred in zip(all_labels, all_preds):
    if true != pred:
        misclass_counts[true] += 1

misclass_rates = misclass_counts / counts  # Normalize by total samples per class

# Print results
for i, label in enumerate(["Black", "Blue", "Green", "TTR"]):
    print(f"{label}: {misclass_counts[i]} misclassifications ({misclass_rates[i]*100:.2f}% of its total samples)")
```

**Observations**:
- Black and TTR have the **highest** misclassification rates (~7–8%).
- Blue and Green are **better classified** (2% or less).
- This aligns with the confusion matrix, which shows **Black ↔ TTR** confusion is common.

Overall, the model is strong, but reducing errors for Black and TTR could further boost the overall accuracy.
